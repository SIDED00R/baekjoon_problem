```python
import math
def prime(x):
  count=0
  for i in range(x+1,x*2+1):
    for j in range(2,int(math.sqrt(i))):
      if i%j!=0:
        count+=1
        break
  return count

a=int(input())
while a!=0:
  print(prime(a))
  a=int(input())

```
